# **Load Data**

```
dataset_dir = 'dataset'
dataset_txt = os.listdir(dataset_dir)
dataset_txt
```

```
panjang = len(dataset_txt)
panjang
```

```
no_putusan = []
informasi_penggugat = []
informasi_tergugat = []
for i in range(0,panjang-1):
  path = "dataset/"+dataset_txt[i]
  # load data
  doc_lower = load_data_to_list_lower(path_doc(path))

  # simpan disclaimer
  index_disclaimer = cari_index_string2(doc_lower,"disclaimer","Disclaimer")
  disclaimer = doc_lower[index_disclaimer:index_disclaimer+4]
    
  #hapus disclamer
  doc_lower = remove_disclamer(doc_lower)

  # hapus halaman
  index_halaman = cari_index_kembar(doc_lower,"halaman")
  for i in index_halaman:
    doc_lower[i] = ""

  # hapus kata kosong
  doc_lower = remove_char(doc_lower,"")

  # no putusan
  no_putusan.append(cari_no_putusan(doc_lower))

  # ekstrak informasi
  informasi_penggugat.append(data_penggugat(doc_lower))
  informasi_tergugat.append(data_tergugat(doc_lower))
```
