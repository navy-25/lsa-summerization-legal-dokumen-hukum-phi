```
email_old = disclaimer[1]
email_old = email_old.split(" ")
email_old = remove_char(email_old,"")

email = []
count = 0 
for i in email_old:
  if "telp" in i:
    start_index_telp = count
    break
  if i == "email" or i == ":":
    continue
  email.append(i)
  count += 1
email = " ".join(email)

telp_old = email_old[start_index_telp+2:]
telp = []
for i in telp_old:
  if "halaman" in i:
    break
  if i == "telp" or i == ":":
    continue
  telp.append(i)
telp = " ".join(telp)
```

```
print("Email   :",email)
print("Telepon :",telp)
```
