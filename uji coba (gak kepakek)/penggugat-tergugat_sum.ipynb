# **Load Data**

```
dataset_dir = 'dataset'
dataset_txt = os.listdir(dataset_dir)
dataset_txt
```

```
panjang = len(dataset_txt)
panjang
```

```
no_putusan = []
informasi_penggugat = []
informasi_tergugat = []
for i in range(0,panjang-1):
  path = "dataset/"+dataset_txt[i]
  # load data
  doc_lower = load_data_to_list_lower(path_doc(path))

  # simpan disclaimer
  index_disclaimer = cari_index_string2(doc_lower,"disclaimer","Disclaimer")
  disclaimer = doc_lower[index_disclaimer:index_disclaimer+4]
    
  #hapus disclamer
  doc_lower = remove_disclamer(doc_lower)

  # hapus halaman
  index_halaman = cari_index_kembar(doc_lower,"halaman")
  for i in index_halaman:
    doc_lower[i] = ""

  # hapus kata kosong
  doc_lower = remove_char(doc_lower,"")

  # no putusan
  no_putusan.append(cari_no_putusan(doc_lower))

  # ekstrak informasi
  informasi_penggugat.append(data_penggugat(doc_lower))
  informasi_tergugat.append(data_tergugat(doc_lower))
```

```
# ekstrak & write data penggugat tergugat di dokumen txt
wr = open("ekstrak data penggugat-tergugat.txt", "w")
wr.write("DATA PENGGUGAT & TERGUGAT\n")
wr.write("==================================================================================================\n")
wr.close()
for i in range(0,len(dataset_txt)-1):
  ap = open("ekstrak data penggugat-tergugat.txt", "a")
  ap.write("Dokumen                          : "+dataset_txt[i]+"\n")
  ap.write("==================================================================================================\n")
  data_penggugat = informasi_penggugat[i]
  ap.write("A. PENGGUGAT\n   - Nama Penggugat              : "+data_penggugat[0]+"\n")
  ap.write("   - Alamat Penggugat            : "+data_penggugat[1]+"\n")
  ap.write("   - Nama Perwakilan Penggugat   : "+data_penggugat[2]+"\n")
  ap.write("   - Alamat Perwakilan Penggugat : "+data_penggugat[3]+"\n")  
  data_tergugat = informasi_tergugat[i]
  ap.write("B. TERGUGAT\n   - Nama Tergugat               : "+data_tergugat[0]+"\n")
  ap.write("   - Alamat Tergugat             : "+data_tergugat[1]+"\n")
  ap.write("   - Nama Perwakilan Tergugat    : "+data_tergugat[2]+"\n")
  ap.write("   - Alamat Perwakilan Tergugat  : "+data_tergugat[3]+"\n")  
  ap.write("==================================================================================================\n")
  ap.close()
```
