# **Function summary gugatan**

```
def summary_gugatan(path_dir,metode):
  # preprocessing
  doc_lower = load_data_to_list_lower(path_doc(path_dir))
  doc_lower = remove_char(doc_lower,"")
  
  start = cari_index_string3(doc_lower,"menimbang,","menimbang, ","menimbang,  ")
  end = cari_index_string3(doc_lower[start:],"menimbang,","menimbang, ","menimbang,  ")
  text_gugatan = doc_lower[start+1:end-1]

  # remove disclammer
  remove_disclamer(text_gugatan)
  # final hasil preprocessing
  text_gugatan = remove_char(text_gugatan,"")
  GUGATAN = text_gugatan

  # Matrix
  from sklearn.feature_extraction.text import CountVectorizer
  vectorizer = CountVectorizer()
  bag_of_words = vectorizer.fit_transform(GUGATAN)

  #Singular value decomposition
  from sklearn.decomposition import TruncatedSVD
  # encode
  svd = TruncatedSVD(n_components = 1)
  decomposition = svd.fit_transform(bag_of_words)

  # normalisasi data
  topic_encoded_df = pd.DataFrame(decomposition, columns=["bobot"])
  topic_encoded_df["doc"]= GUGATAN
  topic_encoded_df["bobot"] = topic_encoded_df["bobot"] / topic_encoded_df["bobot"].max()
  
  # filterisasi
  if metode == "lsa":
    rangkuman=topic_encoded_df[topic_encoded_df["bobot"]>=0.4]
    rangkuman[['doc','bobot']]
  elif metode == "clsa":
    mean = topic_encoded_df["bobot"].mean()
    rangkuman=topic_encoded_df[topic_encoded_df["bobot"]>=mean]
    rangkuman[['doc','bobot']]

  hasil_rangkum = []
  for i in rangkuman['doc']:
    hasil_rangkum.append(i.split())

  # remove bullets & numbering
  remove_bullets_numbering(hasil_rangkum)

  # clean white space hasil lsa
  hasil_rangkum_clean = []
  for i in range(len(hasil_rangkum)):
    hasil_rangkum_clean.append(remove_char(hasil_rangkum[i],""))
  
  # join kata jadi per array
  for i in range(len(hasil_rangkum_clean)):
    hasil_rangkum_clean[i] = " ".join(hasil_rangkum_clean[i])

  # join kalimat menjadi rangkuman
  hasil_rangkum_clean =  " ".join(hasil_rangkum_clean)
  return hasil_rangkum_clean
```
