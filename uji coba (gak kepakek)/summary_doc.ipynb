# **Save hasil rangkuman gugatan ke txt**

```
# LSA
rangkuman_lsa = []
for i in range(len(dataset_txt)):
  path = "dataset/"+dataset_txt[i]
  rangkuman_lsa.append(summary_gugatan(path,"lsa"))

wr = open("summary gugatan lsa.txt", "w")
wr.write("RANGKUMAN GUGATAN LSA\n")
wr.write("==================================================================================================\n")
wr.write("Dokumen   : "+dataset_txt[0]+"\n")
wr.write("Rangkuman :\n"+rangkuman_lsa[0]+"\n")
wr.write("==================================================================================================\n")
wr.close()
for i in range(1,len(rangkuman_lsa)):
  ap = open("summary gugatan lsa.txt", "a")
  ap.write("Dokumen   : "+dataset_txt[i]+"\n")
  ap.write("Rangkuman :\n"+rangkuman_lsa[i]+"\n")
  ap.write("==================================================================================================\n")
  ap.close()
```
